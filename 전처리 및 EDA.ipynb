#### 결측치 열 시각화


```python

# 결측치 시각화 함수
def visualize_missing_data(df):
    # 결측치 비율을 계산
    missing_data = df.isnull().mean() * 100
    
    # 결측치가 0% 이상인 컬럼만 필터링
    missing_data = missing_data[missing_data > 0]
    
    # 결측치 비율을 내림차순으로 정렬
    missing_data = missing_data.sort_values(ascending=False)

    # 한 번에 너무 많은 그래프가 나오지 않도록 쪼개서 출력
    num_plots = len(missing_data)
    cols_per_plot = 5  # 한 플롯에 표시할 열 개수
    num_rows = int(np.ceil(num_plots / cols_per_plot))  # 필요한 행 수 계산 (정수로 변환)
    
    # 그리드 설정
    plt.figure(figsize=(15, num_rows * 3))  # 적절한 크기로 설정
    
    for i, (col, missing) in enumerate(missing_data.items()):
        plt.subplot(num_rows, cols_per_plot, i + 1)
        sns.barplot(x=[col], y=[missing], color='red')  # 결측치 비율을 bar로 표시
        plt.title(f'{col}: {missing:.2f}% Missing')
        plt.ylim(0, 100)  # y축 범위를 0-100으로 설정
    
    plt.tight_layout()  # 그래프 간격 조정
    plt.show()

# 예시 데이터프레임에서 결측치 시각화 호출
visualize_missing_data(df)
```
